```python
import molsysmt as msm
import numpy as np
```

# Translate

```python
molecular_system = msm.demo.classes.metenkephalin(to_form='molsysmt.MolSys')
```

```python
msm.info(molecular_system)
```

```python
msm.get(molecular_system, coordinates=True)
```

```python
n_atoms = msm.get(molecular_system, n_atoms=True)
shifts = np.ones([n_atoms,3], dtype=float)*msm.puw.unit('nm')
```

```python
molecular_system = msm.structure.translate(molecular_system, translation=shifts)
```

```python
coordinates = msm.get(molecular_system, coordinates=True)
coordinates
```

```python
shifts = np.array([-1.0, -1.0, -1.0], dtype=float)*msm.puw.unit('nm')
molecular_system = msm.structure.translate(molecular_system, translation=shifts,
                                           selection=[0,1,2], in_place=False)
```

```python
msm.get(molecular_system, coordinates=True)
```
